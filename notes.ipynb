##### `log(x)` vs `exp(x)` or  `ln(x)`

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of x values for ln(x) and exp(x)
x_ln = np.linspace(0.1, 10, 400)
x_exp = np.linspace(-2, 2, 400)

# Calculate the corresponding y values for both ln(x) and exp(x)
y_ln = np.log(x_ln)
y_exp = np.exp(x_exp)

# Plotting the functions
plt.figure(figsize=(10, 5))

# Plot ln(x)
plt.plot(x_ln, y_ln, label='ln(x) (base e)')

# Plot exp(x)
plt.plot(x_exp, y_exp, label='exp(x)/(e^x)')

# Plot y = x for reference
plt.plot(x_ln, x_ln, 'r--', label='y = x (Reference)')

# Add horizontal and vertical lines at x=0 and y=0
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Title and labels
plt.title('Comparison of ln(x) and exp(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

##### The logistic / sigmoid fuction


**Properties**

1.	Sigmoid Shape: The logistic function has an S-shaped curve.
2.	Asymptotes: It approaches 0 as  x  approaches -\infty and approaches 1 as  x  approaches \infty.
3.	Range: The output values are between 0 and 1.
4.	Inflection Point: The curve is steepest at  x = 0 , where it crosses the midpoint of 0.5.
5.	Derivatives: The first derivative of the logistic function is  f{\prime}(x) = f(x)(1 - f(x)) , showing the rate of change.
6.	Symmetry: The function is symmetric around the inflection point  x = 0 .


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic function

def logistic(h,x,β):
    return h/ (10 + np.exp(-(β*x)))


# Generate x values
x = np.linspace(-10, 10, 400)

################ new!!!! ###########
h= 1 # height
β= 1 # slope

# Generate y values using the logistic function
y = logistic(h,x,β)

# Plot the logistic function
plt.plot(x, y, label='Logistic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Logistic Function')
plt.legend()
plt.grid(True)
plt.show()
```
