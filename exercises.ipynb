```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
data = np.random.randn(1000)

# Create a figure
plt.figure(figsize=(12, 6))

# Create a histogram with density plot using seaborn
sns.kdeplot(data, color='blue', alpha=0.5, label='Histogram with Density')

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with Density Plot using Seaborn')
plt.legend()

# Show the plot
plt.show()
```
